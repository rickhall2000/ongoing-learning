## 8.2 Creating a pivot table from a DataFrame

```python
import pandas as pd

pd.read_csv("sales_by_employee.csv").head()
```

```python
sales = pd.read_csv(
    "sales_by_employee.csv", parse_dates = ["Date"]
)
sales.tail()
```

Steps to create a pviot table:
1. Select the columns you want to aggregate
2. Choose the aggregation opreation to apply
3. Select the columns whose values will group the aggregated data into categories
4. Determine whether to place the groups on the row axis, the column axis, or both axes.

```python
# default aggregation is mean
sales.pivot_table(index="Date")
```

```python
sales.pivot_table(index="Date", aggfunc="sum")
```

```python
sales.pivot_table(index="Date", values="Revenue", aggfunc="sum")
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue",
    aggfunc="sum"
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue",
    aggfunc="sum",
    fill_value=0
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue",
    aggfunc="sum",
    fill_value=0,
    margins=True
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue",
    aggfunc="sum",
    fill_value=0,
    margins=True,
    margins_name="Total"
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue",
    aggfunc="count"
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue",
    aggfunc=["sum", "count"],
    fill_value=0
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values=["Revenue", "Expenses"],
    fill_value=0,
    aggfunc={"Revenue": "min", "Expenses": "max"}
)
```

```python
sales.pivot_table(
    index=["Name", "Date"], values="Revenue", aggfunc="sum"
).head(10)
```

```python
sales.pivot_table(
    index=["Date", "Name"], values="Revenue", aggfunc="sum"
).head(10)
```

## 8.4 Melting a data set

```python
# sales is narrow
sales.head(1)
```

```python
video_game_sales=pd.read_csv("video_game_sales.csv")
video_game_sales.head()
```

```python
# melting converts a wide data set into a narrow one
video_game_sales.melt(id_vars="Name", value_vars="NA").head()
```

```python
regional_sales_columns=["NA", "EU", "JP", "Other"]

video_game_sales.melt(
    id_vars="Name", value_vars=regional_sales_columns
)
```

```python
video_game_sales_by_region = video_game_sales.melt(
    id_vars="Name",
    value_vars=regional_sales_columns,
    var_name="Region",
    value_name="Sales"
)
video_game_sales_by_region.head()
```

```python
# now that we have a narrow representation, it is easier to aggregate
video_game_sales_by_region.pivot_table(
    index="Name", values="Sales", aggfunc="sum"
).head()
```
